```python
company = pd.read_csv('company.csv', parse_dates=['상장일'])
```

```python
company
```

```python
company['상장일'] = pd.to_datetime(company['상장일'])
```

```python
financial_corp = company[company['업종'] == '기타 금융업']
```

```python
kosdaq_financial_corp = financial_corp[financial_corp['분류'] == '코스닥']
```

```python
kosdaq_financial_corp[kosdaq_financial_corp['상장일'] > pd.Timestamp(2015, 1, 1)]
```
